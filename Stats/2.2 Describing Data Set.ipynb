```python
lifetime = pd.read_csv(root_path + 'items-lifetime.csv')
```

```python
lifetime
```

```python
plt.hist(lifetime, bins=10)
plt.show()
```

```python
values = lifetime.iloc[:,0].to_numpy()
lifetime = pd.Series({
    key : np.size(values[(values >= key) & (values < key + 100)])
    for key in range(values.min(), values.max(), 100)
})

lifetime
```

```python
lifetime = lifetime.cumsum() / lifetime.sum()
```

```python
lifetime
```

```python
plt.plot(lifetime.index, lifetime)
plt.xlabel("Item's Lifetime")
plt.ylabel("Relative Cummulative Frequency")
plt.show()
```
