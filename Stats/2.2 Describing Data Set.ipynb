```python
dataset = pd.DataFrame({
    'Salary': [57,58,59,60,61,62,63,64,66,67,70],
    'Frequency': [4,1,3,5,8,10,0,5,2,3,1]
})
```

**Frequency Table** are used to represent a small set of distinct values (need not be numerical)

From the above table we can infer,
- The lowest salary is *57,000*
- The highest salary is *70,000* obtained by one student
- The most common salary is *62,000* obtained by ten students

```python
dataset = pd.DataFrame({
    'Type of Cancer': ['Lung', 'Breast', 'Colon', 'Prostate', 'Melanoma', 'Bladder'],
    'Number of New Cases': [42, 50, 32, 55, 9, 12]
})
```

```python
dataset
```

**Relative Frequency** of a data value is the proportion of the data that have that value.

It can be represented using a pie-chart

```python
dataset['Relative Frequency'] = dataset['Number of New Cases'] \
                                / dataset['Number of New Cases'].sum()
dataset
```

```python
plt.pie(dataset['Relative Frequency'], labels=dataset['Type of Cancer'],
        autopct=lambda freq: f'{round(freq)}%')
plt.show()
```

```python
dataset = pd.read_csv('../Dataset/item-lifetime.csv')
```

```python
dataset
```

For some data set, the number of distinct values is too large in such cases, we divide them into class intervals and then plot the number of data values falling in each classes

- choosing too few classes results in losing too much information
- choosing too high classes results in freq. of each class being too small
- in practise, the class interval is between 5 and 10

```python
intervals = {}

for s in range(500, 1500, 100):
    # We will adopt the left end inclusion convension
    mask = (dataset['Item Lifetime'] >= s) & (dataset['Item Lifetime'] < s + 100)
    intervals[f"{s} - {s + 100}"] = dataset[mask].count()[0]
    
dataset_class = pd.DataFrame(intervals.items(), columns=['Interval', 'Frequency'])
```
